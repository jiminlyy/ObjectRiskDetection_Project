```python
# 셀 2: data.yaml 경로를 Colab 절대경로로 수정
yaml_content = """train: /content/relabeled/train/images
val: /content/relabeled/valid/images

nc: 2
names: ['ped_green', 'ped_red']
"""
with open('/content/relabeled/data.yaml', 'w') as f:
    f.write(yaml_content)
print("data.yaml 수정 완료")
```

```python
# 셀 3: 학습 (런타임을 T4 GPU로 먼저 바꿔놓기)
!pip install ultralytics
from ultralytics import YOLO

model = YOLO("yolov8n.pt")
model.train(
    data="/content/relabeled/data.yaml",
    epochs=50,
    imgsz=640,
    device=0,
    batch=16,
    save_period=10
)
```

```python
# 셀 4: 학습 완료 후 best.pt 다운로드
import glob
weights = glob.glob('/content/runs/detect/train*/weights/best.pt')
weights.sort()
print("찾은 가중치:", weights)
files.download(weights[-1])
```
